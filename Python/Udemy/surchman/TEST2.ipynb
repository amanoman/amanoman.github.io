```python
#ライブラリのインポート  
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# xデータ
x = np.array([25,27,29])

# yデータ
y = np.array([5,21,25])
```

```python
#グラフ描画
plt.scatter(x,y,s=100,c='blue',label= '実データ')
plt.legend(loc='upper left')
plt.title('アイスクリームと気温の関係')
plt.xlabel('最高気温(x)')
plt.ylabel('売れた個数(y)')
plt.show()
```

```python
print(np.mean(x))
```

```python
print(np.mean(y))
```

```python
x_c = x - np.mean(x)
```

```python
y_c = y - np.mean(y)
```

```python
print(x_c)
print(y_c)
```

```python
#データ中心化後のグラフ描画
plt.scatter(x_c,y_c,s=100,c='blue',label= '中心化後の実データ')
plt.legend(loc='upper left')
plt.title('アイスクリームと気温の関係(データ中心化後)')
plt.xlabel('最高気温(x_c)')
plt.ylabel('売れた個数(y_c)')
plt.axhline(0, ls=':', lw=3, c='gray') #lsはLineStyleのこと。'--''=''-.':'などで指定する
plt.axvline(0, ls=':', lw=3, c='gray')
plt.show()
```

```python
#E = a**2 -10a +28 のグラフ 
a = np.arange(0,10,0.1)
E = a**2 -10*a + 28
plt.plot(a,E)
plt.xlabel('a')
plt.ylabel('E')
plt.show()
```

```python
#データの準備
x = np.array([5,10,20,25])
y = np.array([1,2,6,9])

#データのグラフ描画
plt.scatter(x,y,s=100,c='blue',label= '実データ')
plt.legend(loc='upper left')
plt.title('スピードと停止距離の関係')
plt.xlabel('スピード(km/h)')
plt.ylabel('停止距離(m)')
# 近似式
a, b, c = np.polyfit(x, y, 2)
print('a=', a, 'b=',b, 'c=',c )

# 近似式のグラフ描画
x = np.arange(1, 26, 0.1)
y =a * x**2 + b*x + c
plt.plot(x, y,c= 'red')
plt.show()
```
